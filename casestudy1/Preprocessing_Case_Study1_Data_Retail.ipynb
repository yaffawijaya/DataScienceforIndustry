```python
import pandas as pd
```

```python
df = pd.read_csv('Case_Study1_Data_Retail.csv', sep=';', engine='python', header='infer', encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
df_clean = df.dropna()
```

```python
df_clean.info()
```

```python
df_clean['Quantity'] = df_clean['Quantity'].astype(int)
df_clean['Invoice_Date'] = pd.to_datetime(df_clean['Invoice_Date'], format='%d.%m.%Y %H:%M')
df_clean['Price'] = df_clean['Price'].str.replace(',', '.').astype(float)
df_clean['Customer_ID'] = df_clean['Customer_ID'].astype(int)
```

```python
df_clean.info()
```

```python
df_clean
```

```python
df_clean.to_csv(r'clean-Case_Study1_Data_Retail.csv')
```
